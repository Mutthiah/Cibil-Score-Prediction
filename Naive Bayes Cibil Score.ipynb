```python
import pandas as pd
dataset=pd.read_excel("C:\\Users\\mutth\\OneDrive\\Documents\\Custom Office Templates\\cibil score.xlsx")
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
Customer_a=LabelEncoder()
Status_a=LabelEncoder()
AccStatus_a=LabelEncoder()
```

```python
inputs=dataset.drop("Loan",axis="columns")
target=dataset["Loan"]
target
```

```python
inputs["Customer_a"]=Customer_a.fit_transform(inputs["Customer"])
inputs["Status_a"]=Status_a.fit_transform(inputs["Status"])
inputs["AccStatus_a"]=AccStatus_a.fit_transform(inputs["AccStatus"])
inputs
```

```python
inputs2=inputs.drop(["Customer","Year","Status","AccStatus","Score"],axis="columns")
inputs2
```

```python
classifier=GaussianNB()
classifier.fit(inputs2,target)
```

```python
classifier.predict([[5,0,0]])
```

```python
classifier.score(inputs2,target)
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QVBoxLayout, QHBoxLayout, QLineEdit, QPushButton

def get_prediction():
    # Get the user input
    customer_type = int(customer_input.text())
    status = int(status_input.text())
    accstatus = int(accstatus_input.text())
    
    # Make the prediction using the classifier
    prediction = classifier.predict([[customer_type, status, accstatus]])
    
    # Update the prediction label with the result
    prediction_label.setText(f"Prediction: {prediction}")

# ... the rest of the code remains the same until the point where the classifier is created

classifier=GaussianNB()
classifier.fit(inputs2,target)

# Create a QWidget and set a layout for it
app = QApplication(sys.argv)
window = QWidget()
window.setWindowTitle("Loan Prediction")  # Set the window title
layout = QVBoxLayout()

# Create a horizontal layout for the input fields and buttons
input_layout = QHBoxLayout()

# Create labels for the input fields
customer_label = QLabel("Customer Type: ")
input_layout.addWidget(customer_label)
status_label = QLabel("Status: ")
input_layout.addWidget(status_label)
accstatus_label = QLabel("AccStatus: ")
input_layout.addWidget(accstatus_label)

# Create a line edit for the customer type input
customer_input = QLineEdit()
input_layout.addWidget(customer_input)

# Create a line edit for the status input
status_input = QLineEdit()
input_layout.addWidget(status_input)

# Create a line edit for the accstatus input
accstatus_input = QLineEdit()
input_layout.addWidget(accstatus_input)

# Create a push button to get the prediction
get_prediction_button = QPushButton("Get Prediction")
get_prediction_button.clicked.connect(get_prediction)
input_layout.addWidget(get_prediction_button)

# Add the input layout to the main layout
layout.addLayout(input_layout)

# Create a label to display the prediction and add it to the layout
prediction_label = QLabel("Prediction: ")
layout.addWidget(prediction_label)

# Set the layout for the widget and show it
window.setLayout(layout)
window.show()

# Run the application's event loop
sys.exit(app.exec_())
```
